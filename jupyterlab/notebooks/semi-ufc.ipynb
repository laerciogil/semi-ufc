```python
# Example python program to read data from a PostgreSQL table
# and load into a pandas DataFrame
import psycopg2
import pandas as pds
from sqlalchemy import create_engine
```

```python
# Create an engine instance
#alchemyEngine = create_engine('postgresql+psycopg2://test:@127.0.0.1', pool_recycle=3600);

alchemyEngine = create_engine('postgresql://postgres:postgres!@postgres-adventureworks:5432/adventureworks')
```

```python
# Connect to PostgreSQL server
dbConnection = alchemyEngine.connect();
```

```python
# Read data from PostgreSQL database table and load into a DataFrame instance
dataFrame = pds.read_sql("select * from humanresources.department", dbConnection);
```
